```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. c d liu  lu t file CSV
file_path = "C:/bigdata/output/weather_data/part-00000-99b5191a-b78e-4495-ab26-b4625a765b5b-c000.csv"  # iu chnh ng dn nu cn
pandas_df = pd.read_csv(file_path)

# 2. V biu 

##  Biu  Histogram: Phn b nhit 
plt.figure(figsize=(10, 5))
sns.histplot(pandas_df["Temperature (C)"], bins=30, kde=True, color="blue")
plt.title("Phn b nhit  (C)")
plt.xlabel("Nhit  (C)")
plt.ylabel("Tn sut")
plt.grid()
plt.show()

##  Biu  Boxplot: Nhit  theo loi thi tit
plt.figure(figsize=(12, 6))
sns.boxplot(x="Summary", y="Temperature (C)", data=pandas_df)
plt.xticks(rotation=90)  # Xoay nhn  d c
plt.title("Biu  Boxplot - Nhit  theo loi thi tit")
plt.xlabel("Loi thi tit")
plt.ylabel("Nhit  (C)")
plt.show()

##  Biu  Scatter: Quan h gia Nhit  v  m
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Temperature (C)", y="Humidity", data=pandas_df, alpha=0.5)
plt.title("Quan h gia Nhit  v  m")
plt.xlabel("Nhit  (C)")
plt.ylabel(" m")
plt.show()
```
