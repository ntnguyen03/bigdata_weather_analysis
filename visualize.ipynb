```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Đọc dữ liệu đã lưu từ file CSV
file_path = "C:/bigdata/output/weather_data/part-00000-99b5191a-b78e-4495-ab26-b4625a765b5b-c000.csv"  # Điều chỉnh đường dẫn nếu cần
pandas_df = pd.read_csv(file_path)

# 2. Vẽ biểu đồ

## 🔹 Biểu đồ Histogram: Phân bố nhiệt độ
plt.figure(figsize=(10, 5))
sns.histplot(pandas_df["Temperature (C)"], bins=30, kde=True, color="blue")
plt.title("Phân bố nhiệt độ (°C)")
plt.xlabel("Nhiệt độ (°C)")
plt.ylabel("Tần suất")
plt.grid()
plt.show()

## 🔹 Biểu đồ Boxplot: Nhiệt độ theo loại thời tiết
plt.figure(figsize=(12, 6))
sns.boxplot(x="Summary", y="Temperature (C)", data=pandas_df)
plt.xticks(rotation=90)  # Xoay nhãn để dễ đọc
plt.title("Biểu đồ Boxplot - Nhiệt độ theo loại thời tiết")
plt.xlabel("Loại thời tiết")
plt.ylabel("Nhiệt độ (°C)")
plt.show()

## 🔹 Biểu đồ Scatter: Quan hệ giữa Nhiệt độ và Độ ẩm
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Temperature (C)", y="Humidity", data=pandas_df, alpha=0.5)
plt.title("Quan hệ giữa Nhiệt độ và Độ ẩm")
plt.xlabel("Nhiệt độ (°C)")
plt.ylabel("Độ ẩm")
plt.show()
```
